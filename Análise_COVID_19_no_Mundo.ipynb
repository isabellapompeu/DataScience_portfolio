## Obtenção de Dados

A base de dados utilizada nesse artigo foi retirada do site *World in Data*. 

[*World in Data*](https://ourworldindata.org/) é um site que realiza publicações digitais e contém bases de dados abertas e gratuitas. Sua especialidade é expor pesquisas empíricas e dados analíticos sobre mudanças nas condições de qualidade de vida ao redor do mundo.

O dataset se encontra no [GitHub](https://github.com/owid/covid-19-data/tree/master/public/data) e ele é atualizado diariamente.

Como é esperado de uma pandemia, a quantidade e qualidade das informações pode divergir muito ao redor do mundo devido as diferenças entre cultura de dados de cada país. É importante notar que o presente dataset é uma compilação de dados e possui informações faltantes, além disso a instabilidade do cenário global atual também leva à alterações bruscas nos índices de contágio e mortalidade a todo o momento.

**Para iniciarmos nossa análise exploratória, vamos realizar a leitura do nosso arquivo ([nesse link](https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv)), nele vamos encontrar a progressão do coronavírus ao redor do mundo até o dia 20 de agosto de 2020, analisado através de 40 parâmetros diferentes.**

```
# importação dos pacotes necessários
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# leitura do arquivo
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv') ;
```

### Dicionário de Variáveis

Uma descrição completa de todas as colunas do *dataframe* pode ser encontrada [neste link](https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data-codebook.md). No entato, segue-se abaixo as variáveis discutidas nesta análise exploratória:

* `stringency_index` - Government Response Stringency Index: medida composta baseada em 9 indicadores de resposta incluindo fechamento de escolas, fechamento de locais de trabalho e proibição de viagens, redimensionada para um valor de 0 a 100 (100 = resposta mais estrita)

* `gdp_per_capita` - Renda per capta (constante em dólares internacionais 2011), ano mais recente disponível

### O Cenário Atual

Ufa, finalmente vamos realmente começar a analisar nossos dados!

A primeira coisa que faremos é separar apenas a os dados da última data observada, dessa forma teremos um `dataframe` que representa apenas o total de cada varíavel até o momento. A partir daí, podemos começar nossa análise.




```
# criar uma varíavel apenas com a data mais rescente
last_date = df.loc[df.date == '2020-08-20']

# vizualizar a variável
last_date.head()
```

O segundo passo, algo que parece muito intutivo, é descobrir quais foram os países mais e menos afetados pela doença, tanto em número de casos, como de mortes.

Para obtermos uma melhor vizulização, começaremos fazendo gráficos de barras com os *5 países com mais casos de coronavírus* e, desses casos, quantos já resultaram em óbito.

Ao lado, faremos o mesmo processo para os *5 países com menos casos de coronavírus*.

```
# criar varíaveis para o que se deseja vizualisar
top10_cases = last_date.sort_values(by='total_cases', ascending=False).iloc[1:6]
bottom10_cases = last_date.sort_values(by='total_cases', ascending=False).iloc[205:]

# criar figura
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# editar axe 1
ax1.bar('location', 'total_cases', data=top10_cases, width=0.5, color='crimson', label='Total de Casos')
ax1.bar('location', 'total_deaths', data=top10_cases, width=0.5, color='coral', label='Total de Mortes')
ax1.set_title('Top 10 países MAIS afetados')
ax1.set_ylabel('Número de Pessoas (10^6)')
ax1.legend()

# adicionar legendas no topo
rects = ax1.patches

labels = ['* 3º', '* 6º', '* 2º', '* 9º', '* 24º']

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax1.text(rect.get_x() + rect.get_width() / 2, height + 5, label,
            ha='center', va='bottom')
    
# editar axe 2
ax2.bar('location', 'total_cases', data=bottom10_cases, width=0.6, color='crimson', label='Total de Casos')
ax2.bar('location', 'total_deaths', data=bottom10_cases, width=0.6, color='coral', 
        tick_label=['Montserrat', 'Caribbean Netherlands', 'Vatican', 'British Virgin', 'Anguilla'], 
        label='Total de Mortes')
ax2.set_title('Top 10 países MENOS afetados')
ax2.set_ylabel('Número de Pessoas')
ax2.legend()
plt.ylim(top=13)

fig.suptitle('COVID-19 ao Redor do Mundo', y=1.1, fontsize=16)
plt.tight_layout() ;
```

No gráfico à esquerda (Top 10 países MAIS afetados), há uma legenda sobre as barras que indica a posição do respectivo país no ranking populacional de 2020, [neste link](https://www.populationpyramid.net/pt/popula%C3%A7%C3%A3o/2020/). Nesse sentido, percebemos que 4 dos 5 países estão entre os Top 10 países mais populosos *do mundo*, sendo que, o último também está no Top 25. Da mesma forma, quando observamos o gráfico à direita, nenhum dos países possuem populações que passam de 30 mil habitantes. Além disso, quando voltamos ao primeiro gráfico, vemos que, fora os Estados Unidos, todos os outros 4 países se encontram como países emergentes. 

Nota-se, também, que para todos os países apontados (nos dois gráficos), a taxa de mortalidade é baixa, apesar das variações por país.

Esses apontamentos levantam algumas questões:

* O número de casos possui alguma relação direta com a quantidade de habitantes do país ou é apenas um consequência? Já que ao haver uma maior população, há também um maior número de pessoas suscetíveis ao coronavírus.

* Países emergentes ou subdesenvolvidos são mais propensos a possuir um maior número de casos de coronavírus?

Agora que já sabemos quais são os países que mais possuem casos no mundo, vamos ver quais são aqueles onde houve maior número de óbitos e analizar se as mortes são proporcionais ao número de casos.

```
# criar variáveis necessárias
top10_deaths = last_date.sort_values(by='total_deaths', ascending=False).iloc[1:6]

# criar figura
my_colors=['chocolate', 'orangered', 'crimson', 'goldenrod', 'indigo']
plt.figure(figsize=(8.5,5.5))
plt.bar('location', 'total_deaths', data=top10_deaths, width=0.6, color=my_colors)
plt.ylabel('Número de mortes')
plt.title('Maiores números de morte por COVID-19 no mundo', fontdict={'fontsize':16})
plt.show()
```

Apesar dos Estados Unidos e do Brasil se manterem no rank 1 e 2, respectivamente, vemos uma mudança interessante nos outros lugares: México, Índia e Reino Unido; o que mostra que a eficiencia no tratamento da pandemia tem sido muito diferente ao redor do mundo. Além disso, temos uma mudança quando analisamos o número de mortes, pois o número de países *desenvolvidos* aumentou, comparando com os gráficos anteriores gráficos anteriores. 

Países desevolvidos teriam, teoricamente, uma melhor capacidade de tratar seus enfermos e mais recursos para desevolver pesquisas e medidas preventivas, o que nos leva a pensar nos motivos das altas taxas de mortalidade. Algumas das características majoritariamente presentes em países desevolvidos, que podem ter colaborado para uma maior mortalidade seriam: maior expectativa de vida e maior densidade populacional.

Entretanto, precisamos pontuar que existe uma direferença importante entre o número de mortes bruto e sua porcentagem. Em países onde houve um número de casos maior, as chances de haverem mais mortes também é maior (em termos brutos), mas isso não significa, necessáriamente, que a taxa de mortalidade naquele lugar é alta, pelo contrário, mesmo que o índice de recuperação seja muito elevado, países com populações também elevadas podem nos deixar com a *impressão errada*.

É por isso que nosso próximo passo será descobrir quais foram os países com as taxas de mortalidade mais altas. Nós iremos:

* Cacular a média geral de mortalidade por COVID-19;

* Descobrir quais foram os países com taxas de mortalidade superiores a 5%;

* Fazer um histograma com a distribuição de *todas* as taxas de mortalidade;

* E fazer um gráfico de barras comparando os 5 países com as **maiores** taxas de mortalidade.



```
# criação de variáveis necessárias
dropworld = last_date.loc[last_date.location != 'World']
x = dropworld.set_index('location').total_cases
y = dropworld.set_index('location').total_deaths

# cálculo da porcentagem
porcdeaths = ((y / x) * 100).sort_values(ascending=False)

# cálculo da média da porcentagem
(sum(porcdeaths)/len(porcdeaths))
```

**Como nossa média geral é 2.7%, o que confere com índice fornecido pela Organização Mundial da Saúde (que aponta uma méida entre 2 e 4%), nós iremos avaliar países que possuem taxas acima da média, nesse caso, tomaremos 5% como padrão.**

```
# porcentagens superiores a 5%
porcdeaths.loc[porcdeaths > 5]
```

A partir disso, temos que, 28 dos 209 países avaliados apresentaram taxas de mortalidade superiores a 5%, ou seja, mais altas do que a média mundial.

Em pelo menos 7 países a taxa de mortalidade foi superior a 10%, sendo que, no Iêmen esse valor foi superior a 25%. Além disso, mais uma vez podemos ressaltar, **5 dos países do top 7 com as maiores mortalidades, não são apenas países desenvolvidos, mas também, europeus**. 



```
# criar variáveis necessárias
porcdeaths_df = pd.DataFrame(porcdeaths).reset_index(level='location')
porcdeaths_df.columns = ['location', 'porcentage']
porcdeaths_5 = porcdeaths_df.iloc[:5]

# criar figura
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# editar axe 1
ax1.hist(porcdeaths, color='coral')
ax1.set_xlabel('Mortes ( % )')
ax1.set_ylabel('Número de Eventos')
ax1.set_title('Frequência das Mortes')

#editar axe 2
my_colors1=['sienna', 'salmon', 'violet', 'tan', 'yellowgreen']
ax2.bar('location', 'porcentage', data=porcdeaths_5, color=my_colors1, width=0.6)
ax2.set_ylabel('Mortes ( % )')
ax2.set_title('Mortes por país ( % )')
plt.ylim(top=31)

rects2 = ax2.patches
labels2 = ['', '', '', '*', '']

for rect, label in zip(rects2, labels2):
  height2 = rect.get_height()
  ax2.text(rect.get_x() + rect.get_width() / 2, height2, label, ha='center',
           va='bottom')

fig.suptitle('Porcentagem de mortes por COVID-19 no mundo', y=1.1, fontsize=16)

plt.tight_layout();
```

O histograma nos da uma ideia melhor da distribuição da frequência das mortes. Vemos que a maioria dos países apresentou taxas de até 5%, sendo que, aqueles que tiveram mortalidade superior a 7.5% compuseram, na verdade, um *pequeno* grupo de exceções, e é importante lembrar que, de acordo com a lista mostrada antes dos gráficos, boa parte desses países é *desenvolvido*.

O país com a maior taxa de mortalidade no mundo é o Iêmen, onde a situação é realmente castatrófica. Mesmo com o já paupérrimo sistema de saúde local perto do colapso, o único centro dedicado ao tratamente do COVID-19 no sul, em Aden, vem recebendo mais e mais pacientes com chances de tratamento limitadas. O número de enterros multiplicou em 8 vezes desde o início da pandemia, e alguns estudiosos acreditam que as vítimas podem ultrapassar a dos conflitos bélicos na região.

Itália, França, Reino Unido e Bélgica são os próximos países na lista, os três são considerados potências europeias, altamente desenvolvidas e reconhecidos devido a qualidade de vida. Quando comparamos com o número de mortes, o Reino Unido se repete (asterisco acima), o que indica que além de possuir um alto valor absoluto de mortes, também possui uma elevada proporção de mortes por caso.

Nesse sentido, podemos levantar os mesmos questionamentos que foram citados anteriormente: o que levou a uma taxa de mortalidade tão alta em países desenvolvidos? Uma coisa é fato, embora cerca de 90% de toda a população se encontre no Hemisfério Norte, ainda podemos fazer mais investigações, antes de afirmar que foi obra do *acaso*.

Para isso, vamos fazer um quadro de correlações, ou seja, vamos medir o grau de *assossiação* entre nossas variáveis e ver até ponto uma nos dá informações sobre a outra.


```
# criação de variável
dropworld = last_date.loc[last_date.location != 'World']
corr = dropworld.corr()

# cálculo de correlação
fig, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square='True', linecolor='white', annot=True) ;
```

O **coeficiente de correlação de Pearson**, também chamado de "ρ de Pearson", mede o grau da correlação entre duas variáveis de escala métrica. Este coeficiente, normalmente representado por ρ assume apenas valores entre -1 e 1. Quanto maior o valor positivo, maiores as chances de duas variáveis serem *diretamente proporcionais* e, quanto maior o valor negativo, maiores as chances de duas variáveis serem *inversamente proporcionais*.

Possuímos 40 variáveis, ou seja, conseguimos extrair 1600 coeficientes de correlação. Pensando nisso, vamos tentar extrair apenas os principais insights do quadro.

Durante a análise, o parâmetro população do país foi muito questionado a respeito de seu efeito sobre a quantidade de casos e mortes. Na tabela, vemos uma correlação positiva de 0.44 e 0.35 para o total de casos e mortes, respectivamente, indicando que sim, populações maiores tem uma maior tendência a possuir mais casos. No entanto, permanece a dúvida, seria isso uma simples consequência ou um fator ativo?

Outra questão levantada foi a de países *desenvolvidos* apresentarem maiores taxas de mortalidade. Dentre as várias formas de se analisar o desenvolvimento de um país, aqui utilizaremos a `renda per capta` e `instalações para higienização das mãos`, que, no geral, indicam melhores condições de vida.

Nenhuma das variáveis parece possuir altos níveis de correlação direta com os tipos de avaliação do número de casos e de mortes, então, exatamente por que países com capacidades preditivas e sistemas de saneamento básico melhores estão apresentando as maiores taxas de mortalidade? Bom, esses mesmos parâmetros também possuem altos índices de correlação com todos os fatores relacionados a idade (como média geral, pessoas acima de 65 anos, pessoas acima de 70 anos, etc) e como divulgado desde o início do ano por vários veículos de informação, pessoas mais velhas ou comorbidas são mas sensíveis ao novo vírus. Além disso, a `renda per capta` também se relaciona diretamente com características geográficas que podem afetar a eficiência de contágio, como a densidade populacional. 

Nesse sentido é muito provável que tenha ocorrido um *efeito dominó*: países mais desevolvidos, tem maiores expectativas de vida e melhor qualidade de saneamento básico; consequentemente, um maior número de pessoas mais suscetíveis aos sintomas graves do coronavírus (idosos), o que resultou em sistemas de saúde mais obstruídos, e levou a uma precariedade no tratamento da doença.

Esses são apenas alguns fatores que poderiam explicar a maior mortalidade *até o momento*.

E como foi a reação do governo a todo o caos ao redor do mundo? O `stringency_index` (Government Response Stringency Index) vai nos dizer isso; quanto mais alto for ele, mais intensas foram as medidas aplicadas. É irônico ver como ele foi diretamente proporcional a praticamente todos os parâmetros relacionados ao número de casos e número de óbitos, no entanto, essa pode ser simplesmente uma *falsa impressão*, explicada quando observada sua forte correlação com países mais populosos, ou seja, países que majoritáriamente apresetam um maior número de casos.

Um ponto muito interessante é que a quantidade de leitos hospitalares teve correlações muito baixas com a quantidade de mortes por COVID-19, e já quando vemos as instalações para higienização de mãos, essa correlação chegar a ser positiva (fator que pode ser conquência do desenvolvimento do país), o que nos leva a pensar: **o quão importante é a infraestrutura prévia do país, quando comparados a importância de ações governamentais e ao posicionamento da população?**

Por último, vamos analisar os testes. De forma geral, eles nos fazem questionar a capacidade de coleta de dados do COVID-19, uma vez que, para haver casos confirmados, é necessário haver testes e, os testes são feitos, em sua maioria, em países mais desenvolvidos, como por exemplo, com maior `renda per capta`.

Agora que temos uma boa visão geral dos nossos dados, vamos verificar a distribuição dos casos ao longo do ano.

### Série Temporal

A primeira coisa que faremos é verificar a quantidade de casos em cada mês no mundo como um todo, depois disso, vamos analisar as diferenças da progressão da doença em alguns países específicos, tomando como padrão aqueles já surgiram em gráficos anteriores.

No gráfico abaixo, vemos em destaque o número de casos de COVID-19 no mundo ao longo do ano, e logo abaixo, o número de casos de COVID-19 nos 5 países que mais apresentaram casos (valores brutos) e nos 5 países que mais apresentaram mortes por caso.

```
# criação de variáveis
df.date = pd.to_datetime(df.date)
temporal = df.loc[df.date != '2020-08-21']

# criação da figura
fig = plt.figure(constrained_layout=True, figsize=(13, 10))
gs = fig.add_gridspec(nrows=2, ncols=2)

# criar axe 1
ax1 = fig.add_subplot(gs[0, :])
ax1.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'World'])
ax1.legend(['Mundo'])
ax1.set_title('Progressão para o Mundo', fontdict={'size': 14})
ax1.set_ylabel('Número de Casos (10^7)')

# criar axe 2
ax2 = fig.add_subplot(gs[1, 0])
ax2.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'United States'])
ax2.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'Brazil'])
ax2.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'India'])
ax2.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'Russia'])
ax2.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'South Africa'])
ax2.legend(['Estados Unidos', 'Brasil', 'Índia', 'Rússia', 'África do Sul'])
ax2.set_title('Países com mais CASOS no mundo', fontdict={'size': 14})
ax2.set_ylabel('Número de Casos (10^6)')

# criar axe 3
ax3 = fig.add_subplot(gs[1, 1])
ax3.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'Italy'])
ax3.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'France'])
ax3.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'United Kingdom'])
ax3.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'Belgium'])
ax3.plot('date', 'total_cases', data=temporal.loc[temporal.location == 'Hungary'])
ax3.legend(['Itália', 'França',  'Reino Unido', 'Bélgica', 'Hungria'])
ax3.set_title('Países maiores TAXAS DE MORTALIDADE', fontdict={'size': 14})
ax3.set_ylabel('Número de Casos')

fig.suptitle('Progressão dos Casos de COVID-19 ao Longo do Ano', y=1.05, fontsize=16)

plt.tight_layout ;
```

Infelizmente, *nem tudo são flores*, não é mesmo? No último mês, podemos ver que houve um retorno no crescimento de casos nesses países, o que poderia ser a tão famigerada *segunda onda*.


## Conclusão

A capacidade de coletar dados do COVID-19 vem melhorando com o tempo, a medida que os países também melhoram suas infraestruturas e aumentam sua capacidade de fazer testes para o vírus. Esse com certeza é um bom fator, no entanto, pode gerar algumas conclusões erradas, por exemplo, o *surgimento* exagerado de vários casos que nem seriam notados (por serem assintomáticos) em uma situação em que os testes não fossem feitos, ou então, resultados incorretos de testes mais rápidos; e essa é a teoria que dois médicos acadêmicos da Universidade de Stanford, nos Estados Unidos, levataram, propondo que ao dissolver o número de mortes na quantidade de casos que não são notificados, **a taxa de mortalidade por COVID-19 seraia inferior a 2%**.

Ao longo do projeto, vimos também dois perfils principais de países com elevadas taxas de casos e de mortes:

* países subdesenvolvidos ou emergentes, onde a contaminação demorou para se inciar, porém que estão atingindo seus picos nos útimos meses, e, possivelmente por geralmente apresentar maiores populações, também já se encontram em várias posições na lista de países com mais casos e mortes.

* países desenvolvidos, onde, a contaminação começou mais cedo e as taxas de mortalidade são as mais elevadas do mundo. Neles, os índices de contágio também já estavam se estabilizando, até voltarem a começar a subir no mês passando, podendo indicar uma *segunda onda*.

O começo do contágio ter começado mais rápido em países ricos nos mostra que a capacidade de compra e viagens da população influênciou muito na dispersão da doença, como um grande exemplo, há os Estudos Unidos, que como citado, é muito mais longe da China que a Índia por exemplo, mas que começou a ser afetado pelo vírus muito mais cedo.

E por fim, um ponto muito interessante foi notar como as ações do governo apresentaram índices de correlação muito mais altos com as mortes e casos do que a própria a infraestrutura do país, afinal, poucos países possuem ambiente hospitalar para lidar com surtos de doenças súbitos. Isso nos mostra como uma resposta *cirúrgica* do governo pode amenizar tempos de crise.
