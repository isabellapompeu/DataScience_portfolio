## Roubos e Furtos

Quem nunca ouviu uma história de arrastões ou assaltos no Rio de Janeiro não é mesmo? Como foi dito, roubos e furtos enquadram cerca de 30% de todos os crimes reportados no Rio de Janeiro desde 1991.

Primeiramente, qual é a diferença entre roubos e furtos? De acordo com o Código Penal, um **furto** acontece quando não é estabelecido contato com a vítima e pode gerar uma pena de até 4 anos. Já um **roubo** envolve mais violência, uma vez que a vítima está presente no momento do ato e, em decorrência disso, possui penas mais altas, que variam entre 4 e 10 anos.

A partir do nosso *dataset*, vamos analisar alguns dos principais tipos de roubos e furtos no Rio de Janeiro.

### Veículos

No estado, roubos e furtos de veículos compõe 6% de todas as ocorrências registradas desde 1991 pelo banco de dados do Governo do Estado do Rio de Janeiro.

A média de veículos roubados por mês gira em torno de 2448, já a de furtados é de 1675. De acordo com nossa primeira análise estatística, podemos observar também, que houveram meses onde o pico de roubos /furtos duplicou ou triplicou, no entanto, para entendermos melhor a distribuição dos crimes e os períodos de maior criminalidade, faremos uso de representações gráficas:

```
# histograma (variável x contínua)

#criar o histograma
plt.hist([df.roubo_veiculo, df.furto_veiculos], bins=20, label=['Robbery of Vehicle', 'Theft of Vehicle'])

# editar o histograma
plt.legend(loc='upper right')
plt.xlabel('Number of Events')
plt.ylabel('Frequency') ;
```

O gráfico nos permite enxergar que a maior parte do volume de roubos e furtos se concentra entre 1000 e 3000 casos por mês. Entretanto, apesar de dentro dessa faixa de frequencia haver um número maior de furtos, existiram vários momentos com explosões de roubos que ultrapassam 5000 casos, o que justifica sua maior participação no volume total de registros.

Agora, vamos analisar a quantidade de roubos/ furtos em relação ao decorrer do tempo.

```
# somar a quantidade de casos/ ano
roubo = (df.groupby('vano').sum()).iloc[:, 12]
recuperacao = (df.groupby('vano').sum()).iloc[:, 39]
furto = (df.groupby('vano').sum()).iloc[:, 24]
rcoletivo = (df.groupby('vano').sum()).iloc[:, 15]

# criar DataFrame
roubofurto = (pd.concat([roubo, furto], axis=1))

# criar gráfico
roubofurto.plot()
plt.legend(['Robbery of Vehicles', 'Theft of Vehicles'])
plt.xlabel('Year')
plt.ylabel('Number of Events')
```

Podemos ver que, enquanto o número de furtos se manteve quase estável, caindo levemente com o passar do anos, a frequência dos roubos não, sofrendo um pico entre nos anos de 2017 e 2018, passando por uma queda logo após. O que pode indicar que a violência prevalece em "pilhagens" de carros, uma vez que, situações onde o assaltante se encontra diante da vítima são as que dominam.

Também podemos ver, que a partir de 2007, o número de roubos começou a cair e continuou seguindo esse ritmo até voltar a crescer em meados de 2011. É interessante notar que essas datas coincidem com alguns períodos importantes para história do Rio de Janeiro:

* Em 2006 o tráfico de drogas foi reconhecido como crime hediondo pelo Código Penal.

* Em 2008 surge a Unidade de Polícia Pacificadora (UPP), que tinha objetivo acabar com a violência e o domínio territorial do tráfico de drogas.

* Em 2010 desastres naturais assolaram o Rio de Janeiro, deixando vários moradores (que ocupavam áreas de risco) desabrigados devido a deslizamentos de terra.

* Em 2013 o papa faz uma visita ao Rio de Janeiro durante a Jornada Mundial da Juventude.

* Em 2014 acontece a Copa do Mundo no Rio de Janeiro, que gera diversas transformações na cidade e exige novas abordagens para lidar com a violência durante sua preparação (que começou 7 anos antes). Uma nota importante, é que a preparação do Rio de Janeiro foi considerada "A pior que já vi" pelo vice-presidente do Comitê Olímpico Internacional (COI).

* Quedas bruscas do PIB brasileiro desde de 2014.

* Em 2016 acontecem os Jogos Olímpicos no Rio de Janeiro.

#### O que acontece com os carros pilhados?

Bom, essa é uma excelente pergunta! De fato, nossa base de dados não nos permite afirmar quais fins esses veículos tomaram. Contudo, possuimos em mãos a quantidade de veículos posteriormente recuperados, nesse sentido, iremos calcular sua porcentagem em relação ao número de roubos e furtos através da fórmula:

<center> $\frac{\text{recuperacao_veiculos}}{\text{roubo_veiculo} + \text{furto_veiculos}}$ </center>

```
# criação de variáveis
recup = df.recuperacao_veiculos.sum()
roubov = df.roubo_veiculo.sum()
furtov = df.furto_veiculos.sum()

# calculo da formula
tot = ((recup / (roubov + furtov)) * 100).round(2)

# exibicao do resultado
print(f'{tot}% dos carros roubados e furtados são recuperados.')
```

Podemos ver que a taxa de veículos recuperados é muito baixa, sendo inferiror a 50% do volume total. Entretanto, esse valor representa a quantidade total de veículos registrados a partir de 1991, nesse sentido, outros questionamentos são levantados: 'Como os *booms* de roubos e furtos afetam a recuperação de veículos?', 'O desempenho das táticas para recuperação de veículos aumentou ou diminuiu com o passar do tempo?' ou então, 'Existiram momentos onde a taxa de recuperação foi mais alta ou mais baixa que a média?'.

Para respondermos essas questões utilizaremos de outra visualização gráfica.

```
# criar variáveis necessarias
roubo = (df.groupby('vano').sum()).iloc[:, 12]
recuperacao = (df.groupby('vano').sum()).iloc[:, 39]
furto = (df.groupby('vano').sum()).iloc[:, 24]
rcoletivo = (df.groupby('vano').sum()).iloc[:, 15]

# criar DataFrame
veiculos = pd.concat([roubo, recuperacao, furto, rcoletivo], axis=1)

plt.plot(veiculos.index, veiculos.roubo_veiculo, label='Robbery of Vehicles') 
plt.plot(veiculos.index, veiculos.recuperacao_veiculos, label='Recuperation of Vehicles')
plt.plot(veiculos.index, veiculos.furto_veiculos, label='Theft of Vehicles')

# editar gráficos
plt.legend()
plt.xlabel('Year')
plt.ylabel('Number of Events / Month') ;
```

É muito nítido que as apreenssões seguem a curva de roubos ao longo do gráfico, principalmente nos últimos anos. Isso indica que, de forma geral, o número de recuperações aumenta de acordo com o número de roubos e vice-versa, e, acima de tudo, da a entender que a eficiência das operações não foi afetada em decorrência do volume de casos, havendo mais ou menos roubos/ furtos, mas, para confirmar essa afirmação, mais cálculos deveriam ser feitos. Entretanto, para podermos afimar isso com certeza, iremos realizar os cálulos.

Para isso, iremos usar a a quantidade de carros recuperados por ano e dividir pela soma de furtos e roubos por ano. A mesma fórmula de antes, porém, dessa vez, iremos fazer a análise anualmente.

```
# criar variáveis necessarias
roubo = (df.groupby('vano').sum()).iloc[:, 12]
recuperacao = (df.groupby('vano').sum()).iloc[:, 39]
furto = (df.groupby('vano').sum()).iloc[:, 24]

# eficacia da recuperaçao
tot2 = (recuperacao / (roubo + furto)) * 100

# criar gráfico
tot2.plot()
plt.xlabel('Year')
plt.ylabel('Efficiency of Recuperation( % )') ;

```

Apesar de ter havido uma queda na eficácia da recuperação de carros no último, o gráfico nos traz notícias muito positivas, o número de carros recuperados em relação ao númeor de carros pilhados vem aumentando com o passar dos anos e, se compararmos o primeiro e último ano de avaliação, houve uma melhora de cerca de 25% na eficiência da recuperação dos veículos.

## Homicídio Doloso

De acordo com o artigo 18 do Código Penal Brasileiro, um crime doloso é aquele onde o indivíduo quis ou assumiu o resultado. Sendo assim, um homicídio doloso se trata de um crime contra a vida no qual houve intenção de matar, não representando um acidente.

No Rio de Janeiro, a taxa homicídios dolosos é mais de 3 vezes superior a de homicídios culposos, o que pode nos dizer um pouco sobre o perfil da cidade.

A participação de homicídios dolosos dentro de todos os registros feitos desde 1991 é 0.74%, tendo uma média de 504 mortes por mês. Esse tipo de infração possui um elevado desvio padrão, o que pode ser justificado ao analisarmos seus picos (831 mortes) e vales (272 mortes). Além disso, é importante acrescentar que nossa variável não inclui crimes como `lesao_corp_dolosa` ( que possui uma frequencia quase 11 vezes superior).

Uma representação gráfica irá nos permitir visualizar o panorâma dos casos de homicídio doloso no Rio de Janeiro. Para isso, iremos utilizar:

* Um histograma para compreender a distribuição da quantidade de casos.

* Um gráfico de linhas para analisar a frequência dos registros ao longo dos anos.
